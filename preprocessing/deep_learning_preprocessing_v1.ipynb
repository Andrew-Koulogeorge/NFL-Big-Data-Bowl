## Step 0: Pre-Process Play-level data

### A) Filter for only run plays

```python
# Filter for only run plays
run_plays = plays[-plays['playDescription'].str.contains('pass')]
run_plays.head()
```

### B) Feature selection

```python
# Drop extra columns
run_plays_reduced = run_plays.drop(['yardlineSide', 'passResult', 'passLength',
       'penaltyYards', 'playResult', 'playNullifiedByPenalty', 'passProbability', 
       'preSnapHomeTeamWinProbability', 'preSnapVisitorTeamWinProbability', 
       'homeTeamWinProbabilityAdded', 'playDescription',
       'visitorTeamWinProbilityAdded', 'expectedPoints', 'expectedPointsAdded',
       'foulName1', 'foulName2', 'foulNFLId1', 'foulNFLId2'], axis = 1)
run_plays_reduced.head()
```

```python
# Rename outcome variable
run_plays_reduced = run_plays_reduced.rename(columns = {'prePenaltyPlayResult':'TARGET'})
```

```python
# Rename pre-snap variables
pre_snap_vars = ['yardsToGo', 'yardlineNumber', 'gameClock']
for var in pre_snap_vars:
    new_name = "preSnap" + var[0].upper() + var[1:]
    run_plays_reduced = run_plays_reduced.rename(columns = {var:new_name})
run_plays_reduced
```

### C) Preliminary One Hot Encoding

```python
qualitative_vars = ["possessionTeam", "defensiveTeam", "offenseFormation"]
run_plays_ohe = pd.get_dummies(data = run_plays_reduced, columns= qualitative_vars)
print("Old df shape:" + str(run_plays_reduced.shape))
print("New df shape:" + str(run_plays_ohe.shape))
run_plays_ohe
```

```python
run_plays_ohe.columns
```

### D) Join with games data

```python
run_df_clean = games[['gameId', 'week']].merge(run_plays_ohe, on='gameId')
print("pre-join shape: "  + str(run_plays_ohe.shape))
print("post-join shape: " + str(run_df_clean.shape))
run_df_clean.head()
```

## Step 1: Pre-Process tracking data

### A) Drop passing plays (IN PROGRESS)

```python
# Get plays for only week 1
run_play_ids = run_df_clean[run_df_clean['week'] == 1][['gameId','playId']].drop_duplicates()

def drop_non_run_plays(run_play_ids, tracking_df):
    # Merge to filter unique combinations from tracking_df
    # filtered_tracking_df = pd.merge(tracking_df, run_play_ids, on=['gameId', 'playId'], how = 'inner')
    # Boolean indexing to filter unique combinations from tracking_df
    filtered_tracking_df = tracking_df[tracking_df[['gameId', 'playId']].apply(tuple, axis=1).isin(run_play_ids.apply(tuple, axis=1))]
    print("original tracking df shape: " + str(tracking_df.shape))
    print("unique play and game id combos: " + str(run_play_ids.shape))
    print("filtered df shape: " + str(filtered_tracking_df.shape))
    # print("filtered df shape using longer method: " + str(filtered_tracking_df_2.shape))

    # id1 = filtered_tracking_df[['gameId','playId']].drop_duplicates()
    # id2 = filtered_tracking_df_2[['gameId','playId']].drop_duplicates()
    # print(id1.drop_duplicates().equals(id2.drop_duplicates()))
    return filtered_tracking_df
```

```python
filtered_df = drop_non_run_plays(run_play_ids, tracking_week_1)
```

### B) Join tracking, play, and player dataframes

```python
# Methods to link dataframes

def join_play_tracking_data(play_df, tracking_df):
    merged_df = pd.merge(tracking_df, play_df, on=['playId', 'gameId'], how='left')
    print("joined dataframes")
    print("original tracking shape: " + str(tracking_week_1.shape))
    print("merged data shape: " + str(merged_df.shape))
    return merged_df 

def join_player_tracking_data(player_df, tracking_df):
    merged_df = pd.merge(tracking_df, player_df, on=['playerId'], how='left')
    print("original tracking shape: " + str(tracking_week_1.shape))
    print("merged data shape: " + str(merged_df.shape))
    return merged_df
```
